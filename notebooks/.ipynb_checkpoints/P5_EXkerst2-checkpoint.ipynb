## 1. De elektrische veldsterkte $\vec{E}$ beschreven op een x- en y-as

Om in dit xy assenstelsel de vector <span style="color:blue">$\vec{E}$</span> te beschrijven, hebben we twee componenten nodig: <span style="color:red">$E_x$</span> volgens de $\vec{e_x}$ richting en <span style="color:red">$E_y$</span> volgens de $\vec{e_y}$ richting. Zo wordt na een wiskundige afleiding die je in dit project bestudeerde:

$$
\boxed{\vec{E} = E_x \vec{e_x} + E_y \vec{e_y} \quad {\rm met \ als \ componenten} \quad
\begin{cases}
    E_x & = k \, Q \frac{x}{(x^2+y^2)^{3/2}} \\
    E_y & = k \, Q \frac{y}{(x^2+y^2)^{3/2}}
\end{cases}
}
$$

<span style="color:#1569C7">**Opdracht 1.**</span> Schrijf de functie `Exycomponenten(q,xQ,yQ,xP,yP)` die vijf argumenten aanvaardt:
- _q_ is de lading van de bronlading in Coulomb,
- _xQ_ en _yQ_ zijn de x- en y-coordinaten van de bronlading,
- _xP_ en _yP_ de coordinaten zijn van het punt P waar we de veldsterkte willen bepalen.

De retourwaarde van jouw functie zijn de twee componenten $E_x$ en $E_y$.

**Voorbeeld:**

Bij het uitvoeren van jouw functie met argumenten 2.1 nC, 1.0 m, 1.5 m, 4.0 m en 3.0 m:
```python
# maak hier jouw functie Exycomponenten() aan
print( 'E_x,E_y =', Exycomponenten(2.1e-9,1.0,1.5,4.0,3.0), 'N/C')
```
is de uitvoer:
```
E_x,E_y = (1.5009680834566588, 0.7504840417283294) N/C
```

## 3. Voorstelling van de elektrische veldlijnen rond de parallelle platen

<span style="color:#1569C7">**Opdracht 3.**</span> Gebruik nu jouw functies `Exycomponenten(q,xQ,yQ,xP,yP)` en `platen()` uit bovenstaande opdrachten om onderstaande code aan te praat te krijgen. **Als de bovenstaande functies perfect werken, dan is deze opdracht louter een kwestie van de gegeven code uit te voeren** en word je beloond met een mooie voorstelling van het elektrisch veld rond jouw geplaatste platen.

````python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
plt.style.use('dark_background')
import numpy as np

def grid(nx,ny):
    # nx # aantal punten volgens de x-as
    # ny # aantal punten volgens de y-as
    x = np.linspace(-5, 5, nx) # deel de x-as op in nx gelijke stukken binnen de grenzen [-5,5]
    y = np.linspace(-5, 5, ny) # deel de y-as op in ny gelijke stukken binnen de grenzen [-5,5]
    X, Y = np.meshgrid(x, y) # combinatie van de gewenste opsplitsing van de x-as én y-as
    return X,Y

def EGRID(nx,ny):
    #==========================================================#
    #===  BEPAAL DE LADING EN POSITIE VAN DE PUNTLADINGEN   ===#
    #==========================================================#
    charges    = platen() # EX OPDRACHT 1
    #==========================================================#
    #===                   MAAK HET GRID                    ===#
    #==========================================================#
    xP, yP    = grid(nx,ny)
    #==========================================================#
    #=== BEREKEN DE ELEKTRISCHE VELDSTERKTE OP ELK GRIDPUNT ===#
    #==========================================================#
    Ex = np.zeros((ny, nx))
    Ey = np.zeros((ny, nx))
    for charge in charges:
        q, xQ, yQ = charge
        ex, ey = Exycomponenten(q, xQ, yQ, xP, yP) # EX OPDRACHT 2 
        Ex += ex
        Ey += ey
    #==========================================================#
    #== RETOURNEER DE ELEKTRISCHE VELDSTERKTE OP ELK GRIDPUNT =#
    #==========================================================#
    return Ex,Ey


#============================================#
#=== BEREKEN E VOOR ELK PUNT VAN HET GRID ===#
#============================================#
# Voor een goede figuur zorgen we voor voldoende gridpunten (60x60 volstaat)
nx = 60
ny = 60
Ex,Ey = EGRID(nx,ny)

# figuur settings
WIDTH  = 700
HEIGHT = 700
DPI    = 100


#==============#
#=== FIGUUR ===#
#==============#

fig = plt.figure(figsize=(WIDTH/DPI, HEIGHT/DPI))
ax  = fig.add_subplot()
fig.subplots_adjust(left=0, right=1, bottom=0, top=1)

# Plaats de puntladingen (charges in het engels) in het assenstelsel
charges = platen()
for pl in charges:
    q,xco,yco = pl
    if q>0:
        ax.add_artist(Circle([xco,yco], 0.05, color='tab:red', zorder=10))
    else:
        ax.add_artist(Circle([xco,yco], 0.05, color='tab:blue', zorder=10))

# Maak een veldlijnen voorstelling met behulp van de Ex en Ey componenten
color = np.log(np.sqrt(Ex**2 + Ey**2))
ax.streamplot(np.linspace(-5, 5, nx), np.linspace(-5, 5, ny), Ex, Ey, color=color, linewidth=1, cmap=plt.cm.plasma, density=3, arrowstyle='->')

ax.set_xlim(-5,5)
ax.set_ylim(-5,5)
ax.set_aspect('equal')
plt.show()
````
