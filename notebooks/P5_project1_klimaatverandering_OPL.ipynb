Van een string kun je elk afzonderlijk karakter (bv letter in een woord, maar evengoed een leesteken, spatie of cijfer) opvragen met behulp van de vierkante haakjes. Binnen die vierkante haakjes plaatsen we vervolgens het hoeveelste karakter we willen. Vergeet hierbij niet dat Python begint te tellen vanaf 0! Enkele voorbeelden vind je hieronder. Voer de code uit en bekijk het resultaat goed.

```python
a = 'I Love Python'

print( a )
print() #Plaats een witte lijn tussen twee prints
print( 'De string bestaat uit',len(a),'karakters, tel ze maar na!' )
print()
print( a[0],'is het allereerste karakter in de string.' )
print()
print( a[0:6],'zijn de eerste 6 karakters in de string. Het dubbelpunt tussen de vierkante haakjes kun je lezen als "tot".' )
print()
print( a[2:8],'zijn de karakters vanaf het 2e tot het 8e in de string. Merk op dat een spatie ook als een karakter wordt geteld!' )
```

```python
a = '*Dit\tis een voorbeeld-\nstring*'

print( a )
print()
print( 'De string bestaat uit',len(a),'karakters.' )
print()
print( a[0],'is het allereerste karakter in de string.' )
print()
print( a[1],'is het tweede karakter in de string.' )
print()
print( a[1:7],'zijn de karakters vanaf het 1e tot het 7e in de string. Merk op dat de tab als één karakter wordt geteld!' )
# Een tab (= 4 opeenvolgende spaties) kun je in jouw string krijgen door \t te typen
print()
print( a[16:26],'zijn de karakters vanaf het 16e tot het 26e in de string. Merk op dat ook de enter als één karakter wordt geteld!' )
# Een nieuwe lijn kun je in jouw string krijgen door \n te typen
```

### 3.3. Strings splitsen in stukken

Heel leuk, dat selecteren binnen een string, maar de lange string _inhoud_, afkomstig uit het tekstbestand, bevat maar liefst 1976 karakters. Werken met zo'n lange string is heel omslachtig. Om dat te vermijden is het gebruik van de `.split()` methode aan te raden. Met die methode kunnen we een string als argument meegeven die aangeeft waar we de lange string in stukken willen splitsen. Voeren we `a.split(',')` uit op een variabele _a_, dan wordt _a_ gesplitst op elke komma die in _a_ voorkomt. Het resultaat kunnen we in een lijst opslaan. Bijvoorbeeld:

```python
a = '*Dit\tis een voorbeeld-\nstring*'

lijst1 = a.split('v')
print('Gesplitst op v geeft',lijst1)

lijst2 = a.split('i')
print('Gesplitst op i geeft',lijst2)

lijst3 = a.split('\t')
print('Gesplitst op een tab karakter geeft',lijst3)
```

Ook in een lijst kunnen we selecteren op een analoge manier als in strings, met de vierkante haakjes. Bij lijst-selecties selecteren we geen karakter, maar een element. _lijst1_ in het voorbeeld hierboven bestaat uit 2 elementen, gescheiden door een komma in de lijst. Willen we enkel het eerste element uit de lijst, dan selecteren we met behulp van `lijst1[0]` voor het eerste element en `lijst[1]` voor het tweede element. Opmerking: zijn er maar twee elementen en vraag je het derde (niet-bestaande) element op door `lijst[2]`, dan krijg je een **IndexError**:list index out of range.

```python
print( 'Het eerste element uit de eerste lijst is',lijst1[0] )

print( 'Het tweede element uit de eerste lijst is',lijst1[1] )
```

### 3.4. Elementen toevoegen aan een lijst met de .append() methode

<img src="https://files.realpython.com/media/How-to-Use-the-Python-append-Method_Watermarked.4e39826bc6f5.jpg" width="50%">

Een laatste stukje syntax die je nodig hebt om straks de opdracht met succes te kunnen voltooien is de `.append()` methode, die een element toevoegt aan het einde van een bestaande lijst. Als argument binnen de haakjes is er plaats om één element te plaatsen dat je wilt toevoegen. Enkele voorbeelden maken dit hopelijk duidelijk:

```python
q = ['appel','peer','druif'] # de originele lijst met enkele elementen

q.append( 'kiwi' ) # We voegen een extra element aan de lijst toe:

print( 'Na toevoegen van één element krijgen we:',q )

q.append( 'banaan' ) # Nog een extra element

print( 'Na toevoegen van nog een element:',q )
```

```python
mijnlijst = [] # lege lijst met 0 elementen

for i in range(1,11):
    mijnlijst.append(i*7)

print( 'De uiteindelijke lijst is:',mijnlijst )
```

Ik denk dat het principe duidelijk is. Vooral in combinatie in loops, zoals in bovenstaand voorbeeld, worden lijstjes en de `.append()` methode vaak gebruikt. Let ook op de noodzaak van de lege lijst in bovenstaand voorbeeld. Zonder die initieel lege lijst kan de `.append()` methode onmogelijk worden gebruikt. Je kunt geen elementen toevoegen aan iets wat niet bestaat.

Daarnaast moet ik nog op een belangrijk detail wijzen. In feite is het een uitzondering op de regel zoals je die tot nu toe kent: bij om het even welke functie of methode die je kent, is er altijd een invoer als argument binnen de haakjes, en uitvoer die moest worden opgeslagen in een variabele, mogelijk zelfs met dezelfde naam. Overloop maar alle functies die je tot nu toe kent. Echter, de `.append()` methode is wat men in vaktermen '_in place_' noemt: de uitvoer hoef je niet opnieuw op te slaan in een variabele, de variabele waarop je de `.append()` methode toepast, wordt automatisch veranderd!

### 4.2. Scatter-plot
Starten doen we met volgend voorbeeld waarbij we vijf punten in het xy-vlak plaatsen. Losstaande punten in het vlak noemen we in vaktermen een _scatter plot_. Let op volgende lijnen:
- `plt.figure()` voorziet plaats in het geheugen van de computer voor een nieuwe figuur;
- `plt.gca()` is de afkorting voor 'get current axes' en plaatst een xy-assenstelsel op de figuur zoals we die gewoon zijn: met een horizontale x-as en een verticale y-as;
- `plt.show()` toont de figuur. Als er geen extra codelijnen tussen werden gevoegd, verschijnt nu dus een leeg assenstelsel. Uiteraard willen we wel inhoud aan het assenstelsel toevoegen, afhankelijk van het soort figuur verschilt de te gebruiken methode - hier gebruiken we:
- `ax.scatter(x,y)`. De scatter methode wordt toegepast op het assenstelsel (vandaar de ax.scatter()) en verlangt twee argumenten, in het algemeen _x_ en _y_ genoemd, in het voorbeeld hieronder _xpunten_ en _ypunten_. Beide variabelen _xpunten_ en _ypunten_ zijn van het datatype list en bevatten alle x-waarden en y-waarden van de punten. Om de coördinatenkoppels te vormen, moet men dus het eerste element uit de _xpunten_ lijst combineren met het eerste element uit de _ypunten_ lijst, en zo blijven doorgaan tot alle punten zijn geplaatst.

```python
%matplotlib notebook
import matplotlib.pyplot as plt

xpunten = [1,3,5,7,9]
ypunten = [5,3,9,4,7]

# AANMAAK FIGUUR
fig = plt.figure()
ax = plt.gca()
ax.scatter(xpunten,ypunten)
plt.show()
```

### 4.3. Lijn-plot
Veel vaker dan je nu misschien denkt zul je als programmeur echter de verschillende punten willen verbinden met een lijn. Dit kan heel eenvoudig door de `.scatter()` te vervangen door de `.plot()` methode, die verder helemaal gelijk verloopt:

```python
%matplotlib notebook
import matplotlib.pyplot as plt

xpunten = [1,3,5,7,9]
ypunten = [5,3,9,4,7]

# AANMAAK FIGUUR
fig = plt.figure()
ax = plt.gca()
ax.plot(xpunten,ypunten)
plt.show()
```

Hieronder openen we het bestand, lezen de gegevens in met behulp van Python en maken twee lijsten met de gegevens over de massa en temperatuur van de verschillende spectraalklasses:

```python
#===================================#
#===   AANMAAK VAN VARIABELEN    ===#
#===================================#
massas = []
temperaturen = []
lijnteller = 0

#===================================#
#=== OPENEN VAN HET TEKSTBESTAND ===#
#===================================#

with open('P5_project1_bijlage_stereigenschappen.txt','r') as bestand:
    inhoud = bestand.read()

#===================================#
#=== INLEZEN VH GEOPENDE BESTAND ===#
#===================================#

lijnen = inhoud.split('\n') # splits de tabel in lijnen
for lijn in lijnen:
    kolom = lijn.split('\t') # splits de kolommen per lijn
    if lijnteller>0:
        massa = float( kolom[2].replace( ',','.' ) ) # decimaalteken !
        massas.append( massa )
        temp = float( kolom[4] )
        temperaturen.append( temp )
    lijnteller += 1

#===================================#
#===   PRINTEN VAN DE LIJSTEN    ===#
#===================================#

print( massas )
print( temperaturen )
```

```python
import matplotlib.pyplot as plt

#===================================#
#=== AANMAAK VAN DE SCATTERPLOT  ===#
#===================================#

fig = plt.figure()
ax = plt.gca()
ax.scatter(massas,temperaturen,marker='+',color='red',s=100)
ax.set_xlabel('massa (rel tov Zon)')
ax.set_ylabel('temperatuur (K)')
ax.set_title('Verband tussen massa en temperatuur')
plt.show()
```
